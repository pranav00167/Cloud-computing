## 2. Wordcount on Alice's Adventures in Wonderland

Question: How many times does the word `Cheshire` occur?


```python
import re
import ssl
import urllib.request

alice_url = "http://www.gutenberg.org/cache/epub/11/pg11.txt"
ssl_context = ssl._create_unverified_context()

with urllib.request.urlopen(alice_url, context=ssl_context, timeout=30) as response:
    alice_text = response.read().decode("utf-8-sig", errors="replace")

words = re.findall(r"[A-Za-z']+", alice_text)
cheshire_count = sum(1 for word in words if word == "Cheshire")

print("Alice text URL:", alice_url)
print("Count of exact word 'Cheshire':", cheshire_count)
```

**Answer:** `Cheshire` occurs **7** times.

## 3. Median Word Length Using `words.txt`


```python
import re
import ssl
import urllib.request

words_view_url = "https://drive.google.com/file/d/16TIgKhcc2JH8jyJXfwouOV3y7ACX_aas/view?usp=sharing"
ssl_context = ssl._create_unverified_context()

match = re.search(r"/d/([^/]+)/", words_view_url)
if not match:
    raise ValueError("Could not extract file id from the Google Drive link in the PDF.")

file_id = match.group(1)
words_download_url = f"https://drive.google.com/uc?export=download&id={file_id}"

request = urllib.request.Request(words_download_url, headers={"User-Agent": "Mozilla/5.0"})
with urllib.request.urlopen(request, context=ssl_context, timeout=30) as response:
    words_text = response.read().decode("utf-8-sig", errors="replace")

# Hadoop WordMedian uses whitespace tokenization and works with token lengths.
tokens = words_text.split()
lengths = sorted(len(token) for token in tokens)

n = len(lengths)
if n == 0:
    raise ValueError("No tokens found in words.txt")

if n % 2 == 1:
    median_word_length = lengths[n // 2]
else:
    median_word_length = (lengths[n // 2 - 1] + lengths[n // 2]) / 2

print("Original Google Drive link:", words_view_url)
print("Downloaded from:", words_download_url)
print("Number of tokens:", n)
print("Median word length:", median_word_length)
```

**Answer:** The median word length is **4**.
