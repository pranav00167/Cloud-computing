```python

from __future__ import annotations

import csv
import random
from collections import Counter, defaultdict
from pathlib import Path
from typing import Iterable, List, Sequence, Tuple


def create_random_data_file(
    output_file: str = "random_matrix.csv",
    rows: int = 20_000,
    columns: int = 10_000,
    min_value: int = 0,
    max_value: int = 100,
    seed: int = 42,
) -> Path:
    """
    Create a large CSV file using Python's random module.

    Warning:
    The default assignment size is extremely large and may require several GB
    of disk space and a long runtime. The function writes rows incrementally so
    it avoids loading the full matrix into memory.
    """

    random.seed(seed)
    path = Path(output_file)

    with path.open("w", newline="") as csv_file:
        writer = csv.writer(csv_file)
        for _ in range(rows):
            row = [random.randint(min_value, max_value) for _ in range(columns)]
            writer.writerow(row)

    return path


def assign_rows_to_first_row(matrix: Sequence[Sequence[int]]) -> dict[int, List[int]]:
    """
    Assign each subsequent row to the corresponding element in the first row.

    Example:
    first row -> [10, 20, 30]
    second row -> [1, 2, 3]
    third row -> [4, 5, 6]

    Result:
    {
        10: [1, 4],
        20: [2, 5],
        30: [3, 6]
    }
    """

    if not matrix:
        return {}

    first_row = list(matrix[0])
    grouped: dict[int, List[int]] = {value: [] for value in first_row}

    for row in matrix[1:]:
        for index, value in enumerate(row):
            if index < len(first_row):
                grouped[first_row[index]].append(value)

    return grouped


def group_similar_elements(matrix: Sequence[Sequence[int]]) -> dict[int, List[Tuple[int, int]]]:
    """
    Group equal elements and store their positions as (row_index, column_index).
    """

    grouped: dict[int, List[Tuple[int, int]]] = defaultdict(list)

    for row_index, row in enumerate(matrix):
        for column_index, value in enumerate(row):
            grouped[value].append((row_index, column_index))

    return dict(grouped)


def add_matrices(matrix_a: Sequence[Sequence[int]], matrix_b: Sequence[Sequence[int]]) -> List[List[int]]:
    """
    Return the element-wise sum of two matrices.
    """

    validate_same_shape(matrix_a, matrix_b)
    return [
        [value_a + value_b for value_a, value_b in zip(row_a, row_b)]
        for row_a, row_b in zip(matrix_a, matrix_b)
    ]


def subtract_matrices(matrix_a: Sequence[Sequence[int]], matrix_b: Sequence[Sequence[int]]) -> List[List[int]]:
    """
    Return the element-wise difference of two matrices.
    """

    validate_same_shape(matrix_a, matrix_b)
    return [
        [value_a - value_b for value_a, value_b in zip(row_a, row_b)]
        for row_a, row_b in zip(matrix_a, matrix_b)
    ]


def validate_same_shape(matrix_a: Sequence[Sequence[int]], matrix_b: Sequence[Sequence[int]]) -> None:
    if len(matrix_a) != len(matrix_b):
        raise ValueError("Matrices must have the same number of rows.")

    for row_a, row_b in zip(matrix_a, matrix_b):
        if len(row_a) != len(row_b):
            raise ValueError("Matrices must have the same number of columns.")


def create_even_corner_sum_matrix(n: int) -> List[List[int]]:
    """
    Create an n x n matrix of even numbers.

    Using only even numbers guarantees that the sum of opposite corners of
    every 2 x 2 sub-matrix is even.
    """

    value = 2
    matrix = []
    for _ in range(n):
        row = []
        for _ in range(n):
            row.append(value)
            value += 2
        matrix.append(row)
    return matrix


def tuple_matrix_row_wise_addition(
    matrix_a: Sequence[Tuple[int, ...]],
    matrix_b: Sequence[Tuple[int, ...]],
) -> List[Tuple[int, ...]]:
    """
    Perform row-wise tuple matrix addition.
    """

    validate_same_shape(matrix_a, matrix_b)
    result: List[Tuple[int, ...]] = []
    for row_a, row_b in zip(matrix_a, matrix_b):
        result.append(tuple(value_a + value_b for value_a, value_b in zip(row_a, row_b)))
    return result


def print_matrix(matrix: Iterable[Iterable[int]]) -> None:
    for row in matrix:
        print(" ".join(str(value) for value in row))


def demo() -> None:
    sample_matrix = [
        [10, 20, 30],
        [1, 2, 3],
        [4, 5, 6],
        [1, 2, 6],
    ]

    matrix_a = [
        [1, 2, 3],
        [4, 5, 6],
    ]

    matrix_b = [
        [6, 5, 4],
        [3, 2, 1],
    ]

    tuple_matrix_a = ((1, 2), (3, 4))
    tuple_matrix_b = ((5, 6), (7, 8))

    print("\n1. Rows assigned to first-row elements")
    print(assign_rows_to_first_row(sample_matrix))

    print("\n2. Similar elements grouped by position")
    print(group_similar_elements(sample_matrix))

    print("\n3. Matrix addition")
    print_matrix(add_matrices(matrix_a, matrix_b))

    print("\n4. Matrix subtraction")
    print_matrix(subtract_matrices(matrix_a, matrix_b))

    print("\n5. n x n matrix with even opposite-corner sums")
    even_matrix = create_even_corner_sum_matrix(4)
    print_matrix(even_matrix)

    print("\n6. Row-wise element addition in tuple matrix")
    print(tuple_matrix_row_wise_addition(tuple_matrix_a, tuple_matrix_b))

    print("\n7. Large file generator example")
    print(
        "Call create_random_data_file(rows=20000, columns=10000) only if you "
        "have enough disk space and time."
    )


if __name__ == "__main__":
    demo()
```
